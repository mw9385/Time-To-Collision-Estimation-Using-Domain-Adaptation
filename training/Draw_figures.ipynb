```python
# load figures
# scenario (11)
true_11 = np.load('./Results/GAN/CNN/True_11.npy')
# CNN with gan
CNN_gan_mean_11 = np.load('./Results/GAN/CNN/CNN_mean_11.npy')
CNN_gan_unc_11 = np.load('./Results/GAN/CNN/CNN_uncertainty_11.npy')
# ConvLSTM with gan
ConvLSTM_gan_mean_11 = np.load('./Results/GAN/ConvLSTM/ConvLSTM_mean_11.npy')
ConvLSTM_gan_unc_11 = np.load('./Results/GAN/ConvLSTM/ConvLSTM_uncertainty_11.npy')
# CNN without gan
CNN_sim_mean_11 = np.load('./Results/SIM/CNN/CNN_sim_mean_11.npy')
CNN_sim_unc_11 = np.load('./Results/SIM/CNN/CNN_uncertainty_11.npy')
# ConvLSTM without gan
ConvLSTM_sim_mean_11 = np.load('./Results/SIM/ConvLSTM/ConvLSTM_sim_mean_11.npy')
ConvLSTM_sim_unc_11 = np.load('./Results/SIM/ConvLSTM/ConvLSTM_sim_uncertainty_11.npy')
```

```python
# load figures
# scenario (17)
true_17 = np.load('./Results/GAN/CNN/True_17.npy')
# CNN with gan
CNN_gan_mean_17 = np.load('./Results/GAN/CNN/CNN_mean_17.npy')
CNN_gan_unc_17 = np.load('./Results/GAN/CNN/CNN_uncertainty_17.npy')
# ConvLSTM with gan
ConvLSTM_gan_mean_17 = np.load('./Results/GAN/ConvLSTM/ConvLSTM_mean_17.npy')
ConvLSTM_gan_unc_17 = np.load('./Results/GAN/ConvLSTM/ConvLSTM_uncertainty_17.npy')
# CNN without gan
CNN_sim_mean_17 = np.load('./Results/SIM/CNN/CNN_sim_mean_17.npy')
CNN_sim_unc_17 = np.load('./Results/SIM/CNN/CNN_uncertainty_17.npy')
# ConvLSTM without gan
ConvLSTM_sim_mean_17 = np.load('./Results/SIM/ConvLSTM/ConvLSTM_sim_mean_17.npy')
ConvLSTM_sim_unc_17 = np.load('./Results/SIM/ConvLSTM/ConvLSTM_sim_uncertainty_17.npy')
```

```python
n = len(CNN_gan_mean_17)
print(len(CNN_sim_mean_17))
print(n)
```

```python
gan_mean = CNN_gan_mean_17[5:n]
sim_mean = CNN_sim_mean_17[5:n]
gan_unc = CNN_gan_unc_17[5:n]
sim_unc = CNN_sim_unc_17[5:n]
true = true_17[5:n]

fig = plt.figure(figsize=(10,8))
plt.plot(np.arange(n-5), gan_mean, color ='g' ,label = 'GAN')
plt.plot(np.arange(n-5), sim_mean, color = 'orange', label = 'SIM')
plt.plot(np.arange(n-5), true, marker = 'o', color = 'b', label = 'True', alpha = 0.8)

plt.fill_between(np.arange(n-5), gan_mean + gan_unc, gan_mean - gan_unc, alpha = 0.3, color ='g')
plt.fill_between(np.arange(n-5), sim_mean + sim_unc, sim_mean - sim_unc, alpha = 0.3, color = 'orange')

plt.legend(fontsize = '20', loc='lower left')
plt.grid()
plt.ylabel('Time to collision', fontsize = '30')
plt.xlabel('Time Steps', fontsize = '30')
plt.xticks(fontsize=25)    
plt.yticks(fontsize=25)
plt.xlim([0, n-6])
plt.ylim([-1, 7])
plt.show()   
fig.savefig('CNN_scenario_17.png', dpi = 600, bbox_inches = 'tight')
```

```python
n = len(ConvLSTM_gan_mean_17)
print(len(ConvLSTM_sim_mean_17))
print(n)
```

```python
gan_mean = ConvLSTM_gan_mean_17[:n]
sim_mean = ConvLSTM_sim_mean_17[:n]
gan_unc = ConvLSTM_gan_unc_17[:n]
sim_unc = ConvLSTM_sim_unc_17[:n]
true = true_17[5:]

fig = plt.figure(figsize=(10,8))
plt.plot(np.arange(n), gan_mean, color ='g' ,label = 'GAN')
plt.plot(np.arange(n), sim_mean, color = 'orange', label = 'SIM')
plt.plot(np.arange(len(true)), true, marker = 'o', color = 'b', label = 'True', alpha = 0.8)
plt.fill_between(np.arange(n), gan_mean + gan_unc, gan_mean - gan_unc, alpha = 0.3, color ='g')
plt.fill_between(np.arange(n), sim_mean + sim_unc, sim_mean - sim_unc, alpha = 0.3, color = 'orange')

plt.legend(fontsize = '20', loc='lower left')
plt.grid()
plt.ylabel('Time to collision', fontsize = '30')
plt.xlabel('Time Steps', fontsize = '30')
plt.xticks(fontsize=25)    
plt.yticks(fontsize=25)
plt.xlim([0, len(true)-1])
plt.ylim([-1, 6])
plt.show()   
fig.savefig('ConvLSTM_scenario_17.png', dpi = 600, bbox_inches = 'tight')
```
